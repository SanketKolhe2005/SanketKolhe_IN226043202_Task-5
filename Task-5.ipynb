# Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    """
    Calculates total likes and determines post status.
    """
    total_likes = 0
    for like in likes_list:
        total_likes += like

    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")


# Function Call
analyze_post_engagement([200, 300, 400, 150])
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"

    print(f"Average Rainfall: {average}")
    print(f"Rainfall Status: {status}")

# Function call
rainfall_checker([70, 75, 80, 65], 72)
```

# Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    """
    Detects duplicate usernames.
    """
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
        
#Function call
detect_duplicate_accounts(["user1", "user2", "user3", "user1"])
```

# Problem 6: Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(prices):
    """
    Filters crop prices above ₹2000.
    """
    premium = [price for price in prices if price > 2000]

    print(f"Premium Crops: {premium}")

#Function call
filter_premium_crops([1500, 2500, 3200, 1800])
```

# Problem 7: System Monitoring – Application Health Checker

```python
def application_health_checker(error_count):
    """
    Evaluates system health based on error count.
    """
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

#Function call
application_health_checker(7)
```

# Problem 9: E-Learning – Student Attendance Eligibility System

```python
def attendance_checker(attendance_list):
    """
    Calculates attendance percentage and checks eligibility.
    """
    total_classes = len(attendance_list)
    attended = sum(attendance_list)

    percentage = (attended / total_classes) * 100

    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")

#Function call
attendance_checker([1,1,1,0,1,1,1,1,0,1])
```

# Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill_calculator(units):
    """
    Calculates electricity bill and usage classification.
    """

    bill = 0

    # Billing Slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print(f"Total Bill: ₹{bill}")
    print(f"Usage Category: {usage}")

#Function call
electricity_bill_calculator(250)
```
