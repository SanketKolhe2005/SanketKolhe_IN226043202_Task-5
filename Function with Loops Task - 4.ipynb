## Problem 1: Mobile Recharge Validation System

```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        return "Recharge amount must be at least ₹50."
    elif amount not in valid_plans:
        return "Invalid plan selected."
    else:
        return "Recharge Successful!"

# Retry system using while loop
while True:
    try:
        amount = int(input("Enter recharge amount: "))
        result = validate_recharge(amount)
        print(result)
        
        if result == "Recharge Successful!":
            break
    except:
        print("Please enter a valid number.")
```

## Problem 2: Inventory Reorder Alert System

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")

# Sample dictionary
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 20
}

check_inventory(inventory)
```

## Problem 3: Student Result Processing System

```python
# Problem 3: Student Result Processing System

def check_result(marks):
    total = 0
    
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example
student_marks = [60, 70, 45, 80]
result = check_result(student_marks)
print("Result:", result)
```

## Problem 4: Cab Fare Estimator with Retry

```python
# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12
    
    fare = base_fare + (distance * per_km)
    
    if peak == "yes":
        fare += fare * 0.25   # 25% extra
    
    return fare

# Retry using while loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()
        
        total_fare = calculate_fare(distance, peak)
        print("Total Fare: ₹", total_fare)
        break
    except:
        print("Invalid input. Try again.")
```

## Problem 5: Employee Attendance Eligibility Checker

```python
# Problem 5: Employee Attendance Checker

def check_attendance(attendance_list):
    present_days = 0
    
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
attendance = ["P", "P", "A", "P", "P", "A", "P"]
result = check_attendance(attendance)
print("Attendance Status:", result)
```

## Problem 6: Password Strength Checker

```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    if len(password) < 8:
        return "Password too short"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

# Example
user_password = input("Enter password: ")
print(check_password(user_password))
```
